```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
from sklearn import svm
import math
```

## Data Exploration

```python
mpg_data = pd.read_csv("Ex04_Mpg_Regression.csv") 
```

```python
mpg_data
```

```python
mpg_data['carname'] = pd.Categorical(mpg_data.carname).codes
```

```python
mpg_data['horsepower'] = mpg_data['horsepower'].str.replace("?","0.0")
```

```python
mpg_data['horsepower'] = mpg_data['horsepower'].astype(float)
```

```python
mpg_data['horsepower'] = mpg_data['horsepower'].replace(0.0,mpg_data['horsepower'].mean())
```

```python
corr = mpg_data.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
plt.show()
```

### Z Score scaling

```python
mpg_data = (mpg_data - mpg_data.mean())/mpg_data.std()
```

```python
y = np.array(mpg_data['mpg'])
```

```python
l = list(mpg_data.columns)
```

```python
for i in range(1,len(l)):
    x = np.array(mpg_data[l[i]])
    plt.scatter(x,y)
    plt.xlabel(l[i])
    plt.ylabel("mpg")
    plt.show()
```

```python
mpg_data_train_x1 = mpg_data.drop(["mpg","carname","cylinders","acceleration"],axis = 1)
```

```python
mpg_data_train_x = mpg_data_train_x1[:240]
```

```python
mpg_data_train_y = mpg_data['mpg']
```

```python
mpg_data_train_y = mpg_data_train_y[:240]
```

```python
data_x = np.array(mpg_data_train_x)
```

```python
data_y = np.array(mpg_data_train_y)
```

```python
regr = linear_model.Lasso (alpha = 0.2)
```

```python
regr.fit(data_x,data_y)
```

```python
print(mpg_data_train_x.columns)
regr.coef_
```

```python
print(regr.score(data_x,data_y))
```

```python
test_x = np.array(mpg_data_train_x1[241:397])
```

```python
test_x.shape
```

```python
y = regr.predict(test_x)
```
